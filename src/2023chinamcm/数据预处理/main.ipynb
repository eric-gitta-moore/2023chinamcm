```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_excel(r'../数据/附件.xlsx')
df = df.iloc[:, :15]
df = df[df['编号'] != 180]
df = df[df['编号'] <= 390]
df
```

```python
df['整晚睡眠时间（时：分：秒）'] = pd.to_datetime(df['整晚睡眠时间（时：分：秒）'], format='%H:%M:%S').dt.time


def time_to_float(time_obj):
    return time_obj.hour + time_obj.minute / 60 + time_obj.second / 3600


df['整晚睡眠时间（时：分：秒）'] = df['整晚睡眠时间（时：分：秒）'].apply(time_to_float)

df
```

```python
df.dtypes
```

```python
coder = LabelEncoder()
df['婴儿行为特征'] = coder.fit_transform(df['婴儿行为特征'])
df
```

```python
assert isinstance(df, pd.DataFrame)
df.to_excel(r'../数据/预处理后1-390.xlsx', index=False)
```
